```python
import torch

# Model loading
model = torch.hub.load("ultralytics/yolov5", "yolov5s")  # Can be 'yolov5n' - 'yolov5x6', or 'custom'

# Inference on images
img = "https://ultralytics.com/images/zidane.jpg"  # Can be a file, Path, PIL, OpenCV, numpy, or list of images

print(img)  # print image size
# Run inference
results = model(img)

# Display results
results.print()
```

```python
import cv2 as cv
import numpy as np


    

real_height = 15.0  
focal_length = 500  

def calculate_distance(height_in_pixels):

    if height_in_pixels > 0:
        return (real_height * focal_length) / height_in_pixels
    return None

try:
    cam = cv.VideoCapture(0)

    while True:
        check, frame = cam.read()
        if not check:
            break

        results = model(frame)

        for i in range(len(results.xyxy[0])):
            x1, y1, x2, y2, conf, class_id = results.xyxy[0][i]
            height_in_pixels = y2 - y1  

          
            distance = calculate_distance(height_in_pixels)

            
            cv.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            label = f"{model.names[int(class_id)]}: {distance:.2f} cm" if distance else f"{model.names[int(class_id)]}: N/A"
            cv.putText(frame, label, (int(x1), int(y1)), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        cv.imshow("Image", frame)
        key = cv.waitKey(1)
        if key == 27:  
            break

    cam.release()
    cv.destroyAllWindows()
    
except Exception as e:
    print(e)
```
